```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('home-data.csv')
df
```

```python
%matplotlib inline
plt.xlabel('AREA')
plt.ylabel('PRICE')
plt.scatter(df.AREA,df.PRICE,color='red',marker='+')
plt.show()
```

```python
new_df= df.drop('PRICE',axis='columns')
new_df
```

#TRAIN THE MODEL USINGNEW-DF AS FEATURE AND DF.PRICE AS TARGET

```python
reg= linear_model.LinearRegression()
reg.fit(new_df,df.PRICE)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
